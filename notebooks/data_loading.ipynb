## Loading data from GCS bucket

```python
from google.cloud import storage
import tarfile
```

#### Connect with GCS bucket
[Reference: GCP Doc](https://google-cloud-python.readthedocs.io/en/0.9.0/storage-blobs.html)

```python
bucket_name = "dataproc-staging-us-central1-759291875656-wohgf1sk"
storage_client = storage.Client()
bucket = storage_client.bucket(bucket_name)
```

```python
file_prefix = "data/"
blobs = bucket.list_blobs(prefix=file_prefix, delimiter = '/')
```

```python
# list files in folder
for blob in blobs:
    print(blob.name)
```

#### Download file from blob

```python
file_name = "filelist_places365-standard.tar"
blob = bucket.get_blob(file_prefix + file_name)
blob.download_to_filename("file_list.tar")
```

```python
tar_file = tarfile.open("file_list.tar")
arr = []
for member in tar_file.getmembers():
    f = tar_file.extractfile(member)
    if f is not None:
        content = f.read()
        arr.append(content)
tar_file.close()
```

```python
# display an image
from PIL import Image
import matplotlib.pyplot as plt
import io

img_data = arr[0]
image = Image.open(io.BytesIO(img_data))
plt.imshow(image)
```
